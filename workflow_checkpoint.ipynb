```python
from currency_converter import CurrencyConverter
import pandas as pd

c = CurrencyConverter()

def currency_to_EUR(row):
    if row["currency"] != "EUR":
        try:
            conversion = c.convert(row["value"], row["currency"], "EUR")
            return round(conversion, 2)
        except Exception as e:
            print(f"Error converting {row['value']} {row['currency']} to EUR: {e}")
            return None
    else:
        return row["value"]

# Iterate over rows in the DataFrame
for index, row in df.iterrows():
    df.at[index, "value"] = currency_to_EUR(row)
    df.loc[index, "currency"] = "EUR"

# Print rows where the currency is not "EUR"
for index, row in df.iterrows():
    if row["currency"] != "EUR":
        print(row.tolist())
```
